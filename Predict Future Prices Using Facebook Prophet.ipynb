# TASK #3: EXPLORE DATASET  

```python
avocado_df=avocado_df.sort_values('Date') #sorting the dates
```

```python
# Plot date and average price
plt.figure(figsize=(10,10)) #figure function is used to create new figure, figsize(width,height)
plt.plot(avocado_df['Date'],avocado_df['AveragePrice']) #x and y cordinates
```

```python
# Plot distribution of the average price
plt.figure(figsize=(10,6))
sns.distplot(avocado_df['AveragePrice'],color='b') #density distribution
```

```python
# Plot a violin plot of the average price vs. avocado type
sns.violinplot(y='AveragePrice',x='type',data=avocado_df) #plot for different types of avocado
```

```python
# Bar Chart to indicate the number of regions 

sns.set(font_scale=0.9) #xlabels fontsize
plt.figure(figsize=[25,12])
sns.countplot(x = 'region', data = avocado_df) #count of obvs in each categorical bin using bars
plt.xticks(rotation = 45) #xlabels rotation
```

```python
# Bar Chart to indicate the count in every year
sns.set(font_scale=1.5) 
plt.figure(figsize=[25,12])
sns.countplot(x = 'year', data = avocado_df)
plt.xticks(rotation = 45)
```

```python
 # plot the avocado prices vs. regions for conventional avocados
 conventional = sns.catplot('AveragePrice','region',data=avocado_df[avocado_df['type']=='conventional'],
                              hue='year',height=20)
```

```python
  # plot the avocado prices vs. regions for organic avocados
organic=sns.catplot('AveragePrice','region',data=avocado_df[avocado_df['type']=='organic'],hue='year',height=20)
```

# TASK 4: PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado_df
```

```python
avocado_prophet_df=avocado_df[['Date','AveragePrice']]
```

```python
avocado_prophet_df
```

```python
avocado_prophet_df=avocado_prophet_df.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
avocado_prophet_df
```

# TASK 5: UNDERSTAND INTUITION BEHIND FACEBOOK PROPHET

# TASK 6: DEVELOP MODEL AND MAKE PREDICTIONS - PART A

```python
m=Prophet()
m.fit(avocado_prophet_df) #training model with training data
```

```python
# Forcasting into the future
future=m.make_future_dataframe(periods=365)
forecast=m.predict(future)
```

```python
forecast
```

```python
figure=m.plot(forecast,xlabel='Date',ylabel='Price') #light blue area is prediction(trends)
```

```python
figure1=m.plot_components(forecast)
```
